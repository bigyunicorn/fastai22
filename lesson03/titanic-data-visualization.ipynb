The main idea here is that there are three major different types of layer. 
- (1) Input layer — initial data for the neural network.
- (2) Hidden layers — intermediate layer between input and output layer and place where all the computation is done.
- (3) Output layer — produce the result for given inputs.

- Each of the nodes in the input layer represents one input. 
- Each of the nodes in the hidden layers are important as they are the ones who execute computation. 
We can consider each node in the hidden layer looks like this.

(1) Linear regression model. 
The image shows what happens per each row (input). There are total 711 rows, and this process is repeated 711 times with the same coefficients. 

At the end, we sum & average all losses and take gradient descent to update the coefficients to minimize the total loss. 


For neural net, we are going to apply an activation function for each node in a hidden layer and a node in the output layer. 
In the hidden layer, we are using reLu to handle non-linear data. In the output layer, we are simply using sigmoid as the prediction represnts the prediction data. 

THe main difference is that we have two sets of coefficients (weights) as we have two nodes in one hidden layer. 
When we are using the outputs from the hidden layers as the input to the output layer, we do not need to have an additional weight. We simply sum up the two. 

Again, this shows the e2e process of the one input. We have total 711 rows of input. So we are going to calculate the loss by going through the whole training data (711 times) to get the 711 different losess using the same coefficients. 

Then we sum up and average the losses and then take gradient descent to minimize the total loss and update the parameters. 

To minimize the loss more, we can iterate through the entire process for a few times.
